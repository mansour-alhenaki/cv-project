##Objective:

To develop an intelligent computer vision system that monitors workers in industrial environments in real time by detecting worker identity, personal protective equipment (PPE), and unsafe behaviors. The system assigns a unique ID to each worker, tracks their activity, and ensures compliance with safety regulations such as wearing helmets, vests, gloves, goggles, and safety shoes. Additionally, the system analyzes worker interactions with the environment—such as ladder usage—using pose estimation and spatial analysis to detect unsafe actions (e.g., not holding the ladder). When violations occur, alerts are generated and linked to the worker’s ID to enhance workplace safety and reduce accidents.

## **Project Architecture: Factory Safety Monitoring System**

The provided diagram illustrates the hierarchical workflow of our AI-driven safety monitoring system designed for industrial environments. The system operates through a structured six-stage pipeline to ensure worker safety and regulatory compliance.

### **1. Input Stage**
The process begins with the **Input layer**, where real-time data is captured via professional-grade cameras or live video feeds installed throughout the factory floor.

### **2. Detection (The Core Model)**
Using advanced Computer Vision (YOLOv8), the system performs simultaneous multi-class detection. This stage identifies four critical elements:
* **Person Detection:** Locating workers within the frame.
* **PPE Detection:** Identifying safety gear like helmets, vests, and gloves.
* **ID Tag Detection:** Locating the worker's unique identification badge.
* **Ladder Detection:** Monitoring equipment usage and positioning.

### **3. Identification & Tracking**
Once detected, the system moves into the **Identification** phase, where it reads the specific **Worker ID** (e.g., ID 104) to link the data to a specific individual. Simultaneously, the **Tracking** module follows worker movements across the floor to maintain spatial awareness.

### **4. Analysis & Decision**
The **Analysis** engine evaluates the detected data against safety protocols:
* **PPE Compliance:** Does the worker have all required gear?
* **Ladder Risk:** Is the ladder being used safely or placed in a hazardous zone?

Based on this analysis, the system makes an automated **Decision**:
* **Compliant:** Green status (No action needed).
* **Non-Compliant / Danger Present:** Red status (Immediate risk identified).

### **5. Alert System**
In the final stage, if a violation is detected, an **Alert** is instantly dispatched to the supervisor. The notification includes the specific **Worker ID**, the **nature of the violation** (e.g., Missing PPE), and a **Timestamp** for administrative record-keeping.


### **Summary of Tech Stack**
* **Models:** YOLO (You Only Look Once) for real-time object detection.
* **Logic:** Custom Python scripts for compliance checking and ID mapping. 🚨🚨🚨⚠️⚠️⚠️
* **Deployment:** Integrated monitoring dashboard for safety supervisors.

#AI Models Architecture

## Model 1: Object Detection (YOLOv8)
This is the primary vision engine responsible for real-time localization and classification of all critical entities on the factory floor.
* **Personnel:** Human detection (Class: `person`).
* **PPE Compliance:** Detection of safety gear including `helmet`, `vest`, `gloves`, `goggles`, and `boots`.
* **Equipment:** Detection of stationary and mobile `ladder` units.
* **Identification:** Localization of the `ID Tag` on the worker's uniform.

##Model 2: Pose Estimation (YOLOv8-Pose)
This model analyzes the physical orientation and movement of workers to identify hazardous behaviors.
* **Keypoint Mapping:** Tracking of upper limbs (`arms/hands`) and lower limbs (`legs/feet`).
* **Action Recognition:** Used to detect worker-ladder interaction (e.g., climbing, reaching, or improper standing positions).

##Model 3: Optical Character Recognition (OCR)
Once the ID Tag is localized by Model 1, this module extracts the alphanumeric data to identify the individual.
* **Task:** Text extraction and digit recognition from the `ID Tag`.
* **Integration:** Links the visual detection to the worker's profile in the database.

### **Tracking & Re-Identification (DeepSORT)**
To maintain system consistency, we implement the **DeepSORT** (Simple Online and Realtime Tracking with a Deep Association Metric) algorithm.
* **Consistency:** Assigns a **Unique Persistent ID** to every detected worker.
* **Robustness:** Ensures that the worker's ID remains fixed even if they are temporarily occluded or move across different camera views.

---

### **System Workflow Summary**
1.  **YOLOv8** detects the worker and their safety gear.
2.  **DeepSORT** assigns a unique tracking number to the worker.
3.  **OCR** reads the physical ID tag to verify the worker's name/permissions.
4.  **Pose Estimation** monitors the worker's posture for safety violations (e.g., unsafe ladder climbing).

##MODEL 1 : WORKER **ID** TAG VERIFICATION

This model is specifically designed to detect and verify Worker (ID Tags).

It uses a tailored configuration to preserve text readability and ensure high precision.

```python
# ID Modle

#Initialize YOLOv8 Small model (Great balance between speed and accuracy)
model = YOLO("yolov8s.pt")

# Start Training Process
model.train(
    data=id_data,        #Path to the ID dataset YAML fil
    epochs=100,           #Number of complete passes through the dataset
    imgsz=800,           #Input image size
    batch=16,            #Number of images processed at once
    patience=50,         #Stop training if no improvement after 10 epochs
    optimizer="AdamW",   #Modern optimizer for better weight handling
    lr0=0.0001,           #Initial learning rate

    #Data Augmentation Settings (Customized for ID Text)
    degrees=0,           #Minor rotations only to prevent text distortion
    translate=0.05,       #Slight shifting of the image
    scale=0.2,           #Scaling images for different distances
    fliplr=0.0,          #No horizontal flipping to keep text readable

    mosaic=0.0,          #Combine 4 images to help the model see smaller objects
    mixup=0.1,
    cos_lr=True,         #Use Cosine Learning Rate scheduler for smoother convergence
    label_smoothing=0.1, # يساعد الموديل على أن يكون أقل "ثقة عمياء" وأكثر مرونة
    close_mosaic=10,
    plots=True           #Generate charts (Loss, mAP) to monitor performance
)
```

```python
#TEST
import easyocr
import cv2
import os
import random
from ultralytics import YOLO
from google.colab.patches import cv2_imshow

#download the right model
model = YOLO('runs/detect/train/weights/best.pt')

#to read OCR
reader = easyocr.Reader(['en'])

#val
test_images_path = '/content/drive/MyDrive/final_delivery/id/merged_id_dataset/images/val'
all_images = [os.path.join(test_images_path, f) for f in os.listdir(test_images_path)]
random_image = random.choice(all_images) # اختيار صورة عشوائية للتجربة

#تشغيل التوقع
results = model.predict(source=random_image, save=False)

for r in results:
    for box in r.boxes:
        # قص منطقة الـ ID بناءً على تحديد YOLO
        x1, y1, x2, y2 = map(int, box.xyxy[0])
        crop = r.orig_img[y1:y2, x1:x2]

        # تحسين الصورة قليلاً للـ OCR
        gray_crop = cv2.cvtColor(crop, cv2.COLOR_BGR2GRAY)

        #قراءة النص من الجزء المقصوص
        ocr_result = reader.readtext(gray_crop)

        #print
        print(f"--- Detected Box at: [{x1}, {y1}, {x2}, {y2}] ---")
        for res in ocr_result:
            print(f"Detected Text: {res[1]} (Confidence: {res[2]:.2f})")

        cv2_imshow(crop) #عرض الجزء المقصوص عشان تتأكد من دقة YOLO
```

##MODEL 2 : **PPE** COMPLIANCE DETECTION (Personal Protective Equipment)

This model focuses on detecting safety gear: Helmets, Vests, Gloves, etc.

Using YOLOv8m for higher feature extraction capability in complex factory environments.

```python
# PPE Model
#Initialize YOLOv8 Small model
model = YOLO("yolov8s.pt")

# Start Training Process
model.train(
    data= ppe_data ,       #Path to the PPE dataset YAML file
    epochs=100,           #Extended training for better pattern recognition
    imgsz=640,           #High resolution for better small object detection
    batch=16,            #Optimized batch size for GPU memory stability
    patience=15,         #Early stopping if no improvement for 15 epochs
    optimizer="auto",      # Automatically chooses the best optimizer

    #Advanced Training Configuration
    freeze=0,            #Training all layers for full adaptation
    degrees=0,           #No rotation to keep equipment orientation natural
    translate=0.03,      #Minimal shifting
    scale=0.2,           #Scaling images
    fliplr=0.5,          #Safe to flip horizontally for PPE (No text involved)
    mosaic=0.0,          #Disabled to keep the focus on individual object clarity
    cos_lr=True,         #Smooth learning rate reduction
    plots=True           #Enable performance visualization
)
```

```python
import cv2
import os
import random
from ultralytics import YOLO
from google.colab.patches import cv2_imshow

# تحميل مودل PPE
model = YOLO('/content/drive/MyDrive/final_delivery/trained_models/ppe_best.pt')

# مسار صور PPE
test_images_path = '/content/drive/MyDrive/final_delivery/ppe/merged_dataset_v2/images/val'

# اختيار صورة عشوائية
all_images = [
    os.path.join(test_images_path, f)
    for f in os.listdir(test_images_path)
    if f.lower().endswith(('.jpg', '.jpeg', '.png'))
]

random_image = random.choice(all_images)
print("Selected image:", random_image)

# التوقع
results = model.predict(
    source=random_image,
    save=False,
    conf=0.25
)

# عرض النتائج
for r in results:
    annotated = r.plot()
    cv2_imshow(annotated)

    print("\nDetected objects:")
    for box in r.boxes:
        cls_id = int(box.cls[0])
        conf = float(box.conf[0])
        class_name = model.names[cls_id]
        print(f"- {class_name} ({conf:.2f})")
```

## THIRD MODEL: **LADDER** & **EQUIPMENT MONITORING**

This model is designed to detect and monitor ladders within the industrial site.

High augmentation is used to handle different camera angles and distances.


```python
# Ladder Model

#Initialize YOLOv8 Small model (Fast and efficient for static object detection)
model = YOLO("yolov8s.pt")

#Start Training Process
model.train(
    data=ladder_data,    #Path to the Ladder dataset YAML file
    epochs=50,           #Standard epochs for specialized equipment detection
    imgsz=640,           #Standard resolution for object detection
    batch=16,            #Optimized batch size for training stability
    patience=10,         #Stop if mAP doesn't improve for 10 epochs
    optimizer="AdamW",   #Modern optimizer for faster convergence
    lr0=0.001,           #Standard initial learning rate

    #Structural Augmentation (Optimized for Large Objects)
    degrees=10,          #Handle camera tilts and skewed angles
    translate=0.1,       #Slight shifting to simulate different camera placements
    scale=0.5,           #Strong scaling to detect ladders at various distances
    fliplr=0.5,          #Horizontal flip for symmetry

    plots=True           #Visual results for performance analysis
)
```
